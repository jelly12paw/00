```python
import numpy as np
import pandas as pd
import sklearn

from sklearn.tree import DecisionTreeClassifier # 의사결정나무
from sklearn.model_selection import train_test_split # 데이터셋 분류기
from sklearn.metrics import accuracy_score # 모델 성능 측정
from sklearn.metrics import confusion_matrix # 모델 성능 측정 - 오차 행렬
from sklearn.metrics import classification_report # 평가지표 계산
from sklearn.preprocessing import LabelEncoder # 레이블 인코더
from sklearn.neighbors import KNeighborsClassifier # KNN
from sklearn.preprocessing import MinMaxScaler # 최소최대 정규화
from sklearn import svm # 서포트벡터머신
from sklearn.linear_model import LogisticRegression # 로지스틱 회귀분석
from sklearn.ensemble import RandomForestClassifier # 랜덤포레스트 (분류)
```

```python
df = pd.read_csv('http://raw.githubusercontent.com/mwaskom/seaborn-data/master/iris.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
df.groupby('species').size()
```

```python
# 데이터 전처리 : 붓꽃 종류를 숫자로 변경

df['species'].replace({'setosa' : 0, 'versicolor' : 1, 'virginica' : 2 }, inplace = True)

df.groupby('species').size()
```

```python
# 분석 데이터셋 : train_test_split(x, y, test_size = 0.2, random_state = 11)

x = df[['sepal_length', 'sepal_width', 'petal_length', 'petal_width']]
y = df['species']

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 11)

print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

**의사결정나무 (분류)**

```python
# 지도학습 - 의사결정나무 : DecisionTreeClassfier(random_state = 11)

dt = DecisionTreeClassifier(random_state = 11)

dt.fit(x_train, y_train) # 학습 수행

pred = dt.predict(x_test) # 예측 수행

acc = accuracy_score(y_test, pred) # 정확도 측정

print(acc)
```

```python
# 오차 행렬 - confusion_matrix(y_test, pred)

print(confusion_matrix(y_test, pred))
```

```python
# 평가 지표 계산 - classification_report(y_test, pred)

print(classification_report(y_test, pred))
```

```python
df = pd.read_csv('http://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv')
```

```python
df.info()
```

```python
df.head()
```

```python
# 데이터 전처리 - Age 컬럼 결측치를 평균값으로 대체 : fillna()
#                 Embarked 컬럼 결측치를 최빈값으로 대체 : mode()[0]
#                 Cabin 컬럼 제외 : drop()

age_mean = df['Age'].mean()
df['Age'].fillna(age_mean, inplace = True)

em_mode = df['Embarked'].mode()[0]
df['Embarked'].fillna(em_mode, inplace = True)

df.drop(['Cabin'], axis = 1, inplace = True)

df
```

```python
# 데이터 전처리 - 레이블 인코더 : LabelEncoder().fit_transform()
#                 카테고리 컬럼의 값을 숫자 값으로 변환

df['Sex'] = LabelEncoder().fit_transform(df['Sex'])

df['Embarked'] = LabelEncoder().fit_transform(df['Embarked'])

df
```

```python
# 파생변수 추가

df['FamilySize'] = df['SibSp'] + df['Parch']

df
```

```python
# 분석 데이터셋

x = df[['Pclass', 'Sex', 'Age', 'Fare', 'Embarked', 'FamilySize']]
y = df['Survived']

# 분석 데이터셋 분할

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 11)

print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
# 의사결정나무(분류)

dt = DecisionTreeClassifier(random_state = 11)

dt.fit(x_train, y_train)

pred = dt.predict(x_test)

acc = accuracy_score(pred, y_test)

print(acc)
```

```python
df = pd.read_csv('http://raw.githubusercontent.com/mwaskom/seaborn-data/master/iris.csv')
```

```python
# KNN 알고리즘을 이용하여 타이타닉 생존율 예측하기

df = pd.read_csv('http://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv')

age_mean = df['Age'].mean()
df['Age'].fillna(age_mean, inplace = True)

em_mode = df['Embarked'].mode()[0]
df['Embarked'].fillna(em_mode, inplace = True)

df.drop(['Cabin'], axis = 1, inplace = True)

df['Sex'] = LabelEncoder().fit_transform(df['Sex'])

df['Embarked'] = LabelEncoder().fit_transform(df['Embarked'])

df['FamilySize'] = df['SibSp'] + df['Parch']

x = df[['Pclass', 'Sex', 'Age', 'Fare', 'Embarked', 'FamilySize']]
y = df['Survived']

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 11)

knn = KNeighborsClassifier(n_neighbors = 3)
knn.fit(x_train, y_train)

pred = knn.predict(x_test)

acc = accuracy_score(y_test, pred)

print('정확도 :', acc)

# 오차 행렬

mat = confusion_matrix(y_test, pred)
print(mat)

# 평가지표 계산

rpt = classification_report(y_test, pred)
print(rpt)
```

**SVM**

```python
df = pd.read_csv('http://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv')
```

```python
# 데이터 전처리

age_mean = df['Age'].mean()
df['Age'].fillna(age_mean, inplace = True)

em_mode = df['Embarked'].mode()[0]
df['Embarked'].fillna(em_mode, inplace = True)

df.drop(['Cabin'], axis = 1, inplace = True)

df['FamilySize'] = df['SibSp'] + df['Parch']

df.head()
```

```python
# 데이터 전처리 - 원 핫 인코더 : pd.get_dummies()

one_s = pd.get_dummies(df['Sex'])
df = pd.concat([df, one_s], axis = 1)

one_e = pd.get_dummies(df['Embarked'])
df = pd.concat([df, one_e], axis = 1)

df.head()
```

```python
# 훈련셋과 테스트셋 비율을 7 : 3

x = df[['Pclass', 'Age', 'Fare', 'FamilySize', 'female', 'male', 'C', 'Q', 'S']]
y = df['Survived']

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.3, random_state = 10)

print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
# 소프트벡터머신으로 타이타닉 생존율 분석 : svm.SVC(kernel = 'rbf)

sv = svm.SVC(kernel = 'rbf')
sv.fit(x_train, y_train)

pred = sv.predict(x_test)

acc = accuracy_score(y_test, pred)

print(acc)
```

```python
# 모델 성능 평가 - 오차 행렬 / 평가지표 계산

print(confusion_matrix(y_test, pred))

print(classification_report(y_test, pred))
```

```python
# SVM 커널 파라미터 조정

# 커널 파라미터 rbf , C = 0.1, gamma = 0.1
sv = svm.SVC(kernel = 'rbf')

# 커널 파라미터 linear, C = 1, gamma = 0.1
sv = svm.SVC(kernel = 'linear', C = 1, gamma = 0.1)
```

**로지스틱 회귀분석**

```python
# 로지스틱 회귀분석 : LogisticRegression()

lr = LogisticRegression()
lr.fit(x_train, y_train)

pred = lr.predict(x_test)

acc = accuracy_score(y_test, pred)

print(acc)
```

**랜덤포레스트(분류)**

```python
# 랜덤포레스트(분류) : RandomForestClassifier(n_estimator = 50, max_depth = 3, random_state = 20)

rf = RandomForestClassifier(n_estimators = 50, max_depth = 3, random_state = 20)
rf.fit(x_train, y_train)

pred = rf.predict(x_test)

acc = accuracy_score(y_test, pred)

print(acc)
```
